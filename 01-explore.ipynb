# Reading and exploring datasets with *scikit-image* and *pandas*

```python
from matplotlib import pyplot as plt
import pandas as pd
from skimage import io
%matplotlib inline
```

```python
img_fname = '/home/jovyan/datasets/Hydrothermal_Vent_Data/815p5_coax.tif'
img = io.imread(img_fname)
```

```python
dim_x, dim_y = img.shape
print("Width %s, Height %s" % (dim_y, dim_x))
```

```python
plt.figure(figsize=(15,15)) ; plt.imshow(img, cmap='gray');
```

```python
plt.figure(figsize=(10,10)) ; plt.hist(img.reshape(-1), bins=255);
```

```python
img_data = pd.DataFrame(img)
```
